```
from PIL import Image
```

```
def resize(filename, height):
  im = Image.open(f'{filename}')  
  w, h = im.size
  im1 = im.resize((w*60//h,60))
  im1.save(f'{filename}')
```

```
# no gray 808080
name = ['White', 'Gray', 'Black', 'Red', 'Green', 'Blue', 'Yellow', 'Magenta', 'Cyan', 'Orange','Pink', 'Lime', 'Purple','Skyblue']
val =  ['ffffff', '808080', '000000','ff0000', '00ff00', '0000ff', 'ffff00', 'ff00ff','00ffff' , 'ff8000','ff0080','80ff00','8000ff','0080ff']

a = []
for i in range(len(val)):
  r = int(val[i][0:2],16)
  g = int(val[i][2:4],16)
  b = int(val[i][4:],16)
  a.append((r,g,b))
val = a
val
```

```
def dis(c1, c2):
  return ((c1[0]-c2[0])**2+(c1[1]-c2[1])**2+(c1[2]-c2[2])**2)**0.5
```

```
def argmin(a):
    return min(range(len(a)), key=lambda x: a[x])
```

```
def text(file):
  im = Image.open(f'{file}').convert('RGB')
  w, h = im.size
  print(file.split('.')[0] + "=[", end='')

  for j in range(h):
    row = ''
    for i in range(w):
      min_index = argmin([dis(val[k],im.getpixel((i,j))) for k in range(len(val))])
      row += f'{min_index:x}'
    print("'"+row+"',", end='')
  print('],')
```

```
file = 'Pakistan.png'
resize(file, 60)
text(file)
```
